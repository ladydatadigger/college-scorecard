```python
mean_zip = "Resources/MeanZIP-3.csv"
mean_zip_df = pd.read_csv(mean_zip)
mean_zip_df = mean_zip_df.rename(columns={'Zip':'ZIP'})
mean_zip_df["ZIP"]= mean_zip_df["ZIP"].astype(str) 
mean_zip_df.head()
```
