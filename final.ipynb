# Joining All Data with Zipcode Data

```python
#import mean/median household income per zipcode data
mean_zip = "Resources/MeanZIP-3.csv"
mean_zip_df = pd.read_csv(mean_zip)
mean_zip_df = mean_zip_df.rename(columns={'Zip':'ZIP'})
mean_zip_df["ZIP"]= mean_zip_df["ZIP"].astype(str) 
mean_zip_df.head()
```

```python
#import average rental prices per zipcode data
zipcode_prices = "Resources/ZRI_zipcode.csv"
zipcode_df = pd.read_csv(zipcode_prices, encoding='latin-1')
zipcode_df = zipcode_df.rename(columns={'RegionName':'ZIP'})
zipcode_df.head()
```

```python
#merged both zipcode dataframes
zipcode_df["ZIP"]= zipcode_df["ZIP"].astype(str)
zip_joined_df = pd.merge(mean_zip_df, zipcode_df, on='ZIP', how='inner')
zip_joined_df = zip_joined_df[["ZIP", "Median", "Mean", "Zri"]]
zip_joined_df = zip_joined_df.rename(columns={"Median":"Median Household Income in ZIP", "Mean": "Mean Household Income in ZIP", "Zri":"Rental Prices in ZIP"})
zip_joined_df
```

```python
#clean up the zip column so that it is only the first five numbers
start, stop, step = 0, 5, 1
combo_school_df["ZIP"]= combo_school_df["ZIP"].str.slice(start, stop, step) 
combo_school_df
```

```python
#merge joined zipcode data into rest of the college school dataframe
final_df = pd.merge(combo_school_df, zip_joined_df, on='ZIP', how='left')
final_df = final_df[["INSTNM", "CITY", "STABBR", "ZIP", "ADM_RATE_ALL", "TUITIONFEE_IN", "TUITIONFEE_OUT", "MN_EARN_WNE_P10", "MD_EARN_WNE_P10", "GRAD_DEBT_MDN", "Online", "Median Household Income in ZIP", "Mean Household Income in ZIP", "Rental Prices in ZIP"]]
final_df = final_df.rename(columns={"INSTNM":"Institution", "CITY":"City", "STABBR":"State Abbreviation", "ADM_RATE_ALL":"Admissions Rate", "TUITIONFEE_IN":"Tuition In-State", "TUITIONFEE_OUT":"Tuition Out-of-State", "MN_EARN_WNE_P10":"Mean Earnings Post-10 Years", "MD_EARN_WNE_P10": "Median Earnings Post-10 Years", "GRAD_DEBT_MDN":"Median Debt for Post-grads"})
```

```python
final_df.index.name='id'
final_df
```

# Reflect Tables into SQLAlchemy ORM

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/ETL_project')
connection = engine.connect()
```

```python
Base = automap_base()
Base.prepare(engine, reflect = True)
inspector = inspect(engine)

session = Session(engine)
```

```python
final_df.to_sql(name='schools', con=engine, if_exists='append', index=False)
```
