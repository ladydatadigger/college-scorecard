### Web Scraping accredited online schools

```python
url = 'https://www.guidetoonlineschools.com/online-schools?f_show=all'
```

```python
tables = pd.read_html(url)
tables
upper_df=pd.Series
```

```python
len(tables)
```

```python
type(tables)
```

```python
df = tables[4]
#df.columns = ['School', 'Annual Tuition', 'Recommend Rate']
df.head()
```

```python
strip_df = df.columns
strip2_df= df[strip_df].replace({' ':''}, regex = True)
strip2_df.head()
```

```python
strip2_df["School"]=strip2_df["School"].apply(str.upper)
strip2_df.head()
```

```python
online_schools_df = strip2_df.drop(columns =["Annual Tuition", "Recommend Rate"])
online_schools_df.head()
```

```python
online_schools_df['School'] =online_schools_df['School'].str.replace("&","")
online_schools_df['School'] =online_schools_df['School'].str.replace("-","")
online_schools_df['School'] =online_schools_df['School'].str.replace(".","")
```

```python
online_schools_df['Online'] = "TRUE"
online_schools_df.head()
```

```python
online_schools_df = online_schools_df.rename(columns={"School":"UP_INSTNM"})
online_schools_df.head()
```

```python
online_schools_df.set_index("UP_INSTNM")
```

```python
online_schools_df.to_csv('Resources/Online-schools.csv')
```

### Joining Online Accredited School with All-Data 

```python
all_data = "Resources/All-Data.csv"
all_data_df = pd.read_csv(all_data)
```

```python
#Add column of institution name (Strings only)
all_data_df['UP_INSTNM'] = all_data_df['INSTNM'] 
```

```python
all_data_df['UP_INSTNM'] = all_data_df['UP_INSTNM'].str.upper() 
all_data_df['UP_INSTNM'] = all_data_df['UP_INSTNM'].str.replace(" ","")
all_data_df['UP_INSTNM'] = all_data_df['UP_INSTNM'].str.replace("-","")
all_data_df['UP_INSTNM'] = all_data_df['UP_INSTNM'].str.replace("&","")
all_data_df['UP_INSTNM'] = all_data_df['UP_INSTNM'].str.replace(".","")
all_data_df.head()
# print(all_data_df['UP_INSTNM'])
```

```python
# join all-data mod and online_schools_df
combo_school_df = pd.merge(all_data_df, online_schools_df, on="UP_INSTNM", how="left")

combo_school_df
```

```python
UPINSTN=[]
for index,row in combo_school_df.iterrows():
    if row["Online"]=="TRUE":
        UPINSTN.append(row["UP_INSTNM"])
        print(row["UP_INSTNM"])
```
